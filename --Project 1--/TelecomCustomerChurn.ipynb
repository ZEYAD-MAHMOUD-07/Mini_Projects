```python
###### Cleaning Data ######
```

```python
df.dropna(axis='index',how='any',subset=['PaymentMethod'],inplace = True)
```

```python
df.info()
```

```python
df['Tenure_missing'] = df['Tenure'].isna().astype(int)
df['Tenure'] = df['Tenure'].fillna(0)
```

```python
df.info()
```

```python
df.describe(include='number')
```

```python
txt_cols = df.select_dtypes(include='object').columns
```

```python
for col in txt_cols:
    df[col]=df[col].str.strip().str.lower()
```

```python
df.head(10)
```

```python
binary_cols = [col for col in txt_cols
                  if df[col].isin(['yes','no']).all()]

for col in binary_cols:
    df[col]=df[col].map({'yes':1,'no':0})
```

```python
df.head()
```

```python
df['Gender']=df['Gender'].map({'male':1,'female':0})
df['Gender']
```

```python
churned_after_18month = df[(df['Churn']==1)&(df['Tenure']>=18)].shape[0]
churned_before_18month = df[(df['Churn']==1)&(df['Tenure']<18)].shape[0]

values = np.array([churned_after_18month,churned_before_18month])
lbls = np.array(['churned_after_18month','churned_before_18month'])
```

```python
plt.pie(x=values,labels=lbls,shadow=True,startangle=90,autopct="%1.2f",explode=[0.045,0.005])
plt.title("Customers churn before and after one and half year",fontweight='bold')
```
